```python
import pandas as pd
import sys
sys.path.append('..')
from pipeline.preprocessors.revisedMissing import RevisedMissingValueHandler
from pipeline.preprocessors.numerical import Scaler
```

```python
df = pd.read_csv('../../data/diabetes.csv')
```

```python
impossibles = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']
missing_handler = RevisedMissingValueHandler(
    drop_cols=['Insulin'],
    impute_cols=impossibles,
    zero_cols=impossibles
)

df_clean = missing_handler.fit_transform(df)
```

```python
scaler = Scaler(scale_cols=['Glucose', 'BloodPressure', 'SkinThickness',
                             'BMI', 'Pregnancies', 'DiabetesPedigreeFunction', 'Age'])
df_scaled = scaler.fit_transform(df_clean)
df_scaled.head()
```
